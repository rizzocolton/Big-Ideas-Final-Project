Importing pandas first, grabbing csv and sorting it by neighborhood (alphabetically).

```python
import pandas as pd

properties=pd.read_csv("Condemned-Properties.csv")
properties.head(10)
```

Creating barchart to visualize data, showing each neighborhood's # of condemned/dead-end properties

```python
prop_counts=properties['neighborhood'].value_counts()
prop_counts.plot(kind='bar',figsize=(24,8),rot=80)
```

Top 10 lowest condemned property counts

```python
low_condemnation_neighborhoods=prop_counts.tail(10)
low_condemnation_neighborhoods.head(10)
```

Assigning scores to each prop (1 = lowest # of condemned properties, 0 = highest # of condemned properties)

```python
range=prop_counts.max()-prop_counts.min()
min=prop_counts.min()
prop_scores=pd.Series()
for prop, value in prop_counts.items():
    prop_scores.at[prop]=1-(value-min)/range
prop_scores.tail(10)
prop_scores.plot(kind='bar',figsize=(24,8),rot=80)
```
